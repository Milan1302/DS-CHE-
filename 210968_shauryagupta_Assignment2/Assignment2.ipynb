```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
rng = np.random.RandomState(10)
x = 10 * rng.rand(50)
y = 2 * x - 5 + rng.randn(50)
plt.scatter(x, y);
```

```python
type(y)
```

```python
def loss(y_hat, y):
    l = ((abs(y_hat-y)**3)+abs(y_hat-y))/2
    return l.mean()
```

```python
x.shape
```

```python
L=[]
N=x.shape[0]
```

```python
N
```

```python
def forward_pass(x,m,c):
    return m*x+c
```

```python
m = np.random.randn(1)
m
```

```python

m1 = np.random.randn(1)
c1 = np.random.randn(1)


learningrate = 0.01
numiter = 50000

for i in range(numiter):
   
    y_hat=forward_pass(x,m1,c1)
    
    dm = (2.0/N) *  -np.sum(np.multiply(x,(y - (m1*x +c1))))
   
    dc = (2.0/N) * -np.sum(y - (m1*x + c1))
    
    m1 = m1 -learningrate * dm
    c1 = c1 - learningrate * dm
    if i%5000==0:
        L.append(loss(y_hat,y))
        print ("no of iterations:",i )
        plt.plot(x,y,'go')
        plt.plot(x,y_hat,'r*')
    
        plt.ylabel('Data')
        plt.show()
```

```python
plt.plot(L)
```

```python
def loss2(y_hat, y):
    l = ((abs(y_hat-y)**4)+(abs(y_hat-y)**7))/2
    return l.mean()
```

```python
def forward_pass2(x,a,b,c):
    return a*(x**2) + b*x + c
```

```python
a1 = np.random.randn(1)
b1 = np.random.randn(1)
c1 = np.random.randn(1)


learningrate = 0.0001
numiter = 50000
for i in range(numiter):
   
    y_hat=forward_pass2(x,a1,b1,c1)
    
    da = (2.0/N) *  -np.sum(np.multiply(x**2,(y - (a1*(x**2) + b1*x + c1))))
   
    db = (2.0/N) *  -np.sum(np.multiply(x,(y - (a1*(x**2) + b1*x + c1))))
    
    dc = (2.0/N) * -np.sum(y - (a1*(x**2) + b1*x + c1))
    
    a1 = a1 -learningrate * da
    b1 = a1 -learningrate * db
    c1 = a1 -learningrate * dc
    if i%5000==0:
        L.append(loss2(y_hat,y))
        print ("no of iterations:",i )
        plt.plot(x,y,'go')
        plt.plot(x,y_hat,'r*')
    
        plt.ylabel('Data')
        plt.show()
```
