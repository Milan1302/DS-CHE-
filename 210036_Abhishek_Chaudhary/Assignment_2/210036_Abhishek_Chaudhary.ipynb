```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
rng = np.random.RandomState(10)
x = 10 * rng.rand(50)
y = 2 * x - 5 + rng.randn(50)
plt.scatter(x, y);
```

```python
# Define the loss functions
def mean_cubed_error(x, x̂):
    return np.mean(np.abs(x - x̂)**3)

def mean_abs_error(x, x̂):
    return np.mean(np.abs(x - x̂))

def mean_quartic_error(x, x̂):
    return np.mean((x - x̂)**4)

def mean_septic_error(x, x̂):
    return np.mean((x - x̂)**7)
```

```python
L=[]
N=x.shape[0]
```

```python
def forward_pass(x,m,c):
    return m*x+c
```

```python
# Random Initialisation
m1 = np.random.randn(1)
c1 = np.random.randn(1)

# Hyperparameters
learningrate = 0.01
numiter = 50000

for i in range(numiter):
    # forward pass
    y_hat=forward_pass(x,m1,c1)
    # Loss
   
    # Gradient
    dm = (-3/N) * np.mean(y - y_hat) * (y - y_hat)**2
    dc = (-3/N) * np.mean(x * (y - y_hat)**2)
    
    m1 = m1 -learningrate * dm
    c1 = c1 - learningrate * dm
    if i%5000==0:
        L.append(mean_cubed_error(y_hat,y))
        print ("no of iterations:",i )
        plt.plot(x,y,'go')
        plt.plot(x,y_hat,'r*')
    #plt.plot(x,m1*x+c1,'r*')
        plt.ylabel('Data')
        plt.show()
```

```python
plt.plot(L)
```

```python
rng = np.random.RandomState()
x_q = 10 * rng.rand(50)
y_q = x_q**2 - 5*x_q + 6 + rng.randn(50)
plt.scatter(x_q, y_q);
```

```python
def forward_pass2(x,a,b,c):
    return a*x**2 + b*x + c
```

```python
# Random Initialisation
a1 = np.random.randn(1)
b1 = np.random.randn(1)
c1 = np.random.randn(1)

# Hyperparameters
learningrate = 0.000000001
numiter = 50000

for i in range(numiter):
    # forward pass
    y_hat3=forward_pass2(x_q,a1,b1,c1)
    # Loss
   
    # Gradient
    da =  (-4/N) * np.mean(x**2 * (y - y_hat3)**3)
    db = (-4/N) * np.mean(x * (y - y_hat3)**3)
    dc =  (-4/N) * np.mean((y - y_hat3)**3)
    
    a1 = a1 - learningrate * da
    b1 = b1 - learningrate * db
    c1 = c1 - learningrate * dc

    if i%5000==0:
        L.append(mean_quartic_error(y_hat3,y_q))
        print ("no of iterations:",i )
        plt.plot(x_q,y_q,'go')
        plt.plot(x_q,y_hat3,'r*')
    #plt.plot(x,m1*x+c1,'r*')
        plt.ylabel('Data')
        plt.show()
```
