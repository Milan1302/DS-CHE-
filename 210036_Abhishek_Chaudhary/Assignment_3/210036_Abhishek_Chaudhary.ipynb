```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('superheated_vapor_properties.csv')
df
```

```python
df1 = df.loc[(df["Property"] == "V"),:]
df1
```

```python
def R2_score(y, y_hat):
    y_mean = np.mean(y)
    x = np.sum((y - y_mean)**2)
    y = np.sum((y - y_hat)**2)
    return 1 - y / x
```

```python
df_1 = df1.loc[(df1["Pressure"] < 300),:]
df_2 = df1.loc[(df1["Pressure"] >= 300) & (df1["Pressure"] < 1500),:]
df_3 = df1.loc[(df1["Pressure"] >= 1500),:]
```

```python
plt.plot(df_1['Pressure'],df_1['Vap_Sat'])
```

```python
plt.plot(df_2['Pressure'],df_2['Vap_Sat'])
```

```python
plt.plot(df_3['Pressure'],df_3['Vap_Sat'])
```

```python
P_1 = df_1['Pressure'].values.reshape(-1,1)
Vap_sat_1 = df_1['Vap_Sat'].values.reshape(-1,1)
P_2 = df_2['Pressure'].values.reshape(-1,1)
Vap_sat_2 = df_2['Vap_Sat'].values.reshape(-1,1)
P_3 = df_3['Pressure'].values.reshape(-1,1)
Vap_sat_3 = df_3['Vap_Sat'].values.reshape(-1,1)
```

```python
def loss(y_hat, y):
    loss = (y_hat - y) ** 2 / 2
    return loss.mean()
```

```python
L_1=[]
L_2=[]
L_3=[]
N_1=P_1.shape[0]
N_2=P_2.shape[0]
N_3=P_3.shape[0]
```

```python
def forward_pass(x,m,c):
    return m*x+c
```

```python
# Random Initialisation
m1 = np.random.randn(1)
c1 = np.random.randn(1)

# Hyperparameters
learningrate = 1e-6
numiter = 50000

for i in range(numiter):
    # forward pass
    y_hat=forward_pass(P_1,m1,c1)
    # Loss
   
    # Gradient
    dm = (1.0/N_1) *  -np.sum(np.multiply(P_1,(Vap_sat_1 - (m1*P_1 +c1))))
   
    dc = (1.0/N_1) * -np.sum(Vap_sat_1 - (m1*P_1 + c1))
    
    m1 = m1 -learningrate * dm
    c1 = c1 - learningrate * dc
    if i%5000==0:
        L_1.append(loss(y_hat,Vap_sat_1))
        print ("no of iterations:",i )
        plt.plot(P_1,Vap_sat_1,'go')
        plt.plot(P_1,y_hat,'r*')
        plt.ylabel('Data')
        plt.show()
        
R_1 = R2_score(P_1, y_hat)
```

```python
# Random Initialisation
m2 = np.random.randn(1)
c2 = np.random.randn(1)

# Hyperparameters
learningrate = 1e-6
numiter = 50000

for i in range(numiter):
    # forward pass
    y_hat=forward_pass(P_2,m2,c2)
    # Loss
   
    # Gradient
    dm = (1.0/N_2) *  -np.sum(np.multiply(P_2,(Vap_sat_2 - (m2*P_2 +c2))))
   
    dc = (1.0/N_2) * -np.sum(Vap_sat_2 - (m2*P_2 + c2))
    
    m2 = m2 -learningrate * dm
    c2 = c2 - learningrate * dc
    if i%5000==0:
        L_2.append(loss(y_hat,Vap_sat_2))
        print ("no of iterations:",i )
        plt.plot(P_2,Vap_sat_2,'go')
        plt.plot(P_2,y_hat,'r*')
        plt.ylabel('Data')
        plt.show()
        
R_2 = R2_score(P_2, y_hat)
```

```python
# Random Initialisation
m3 = np.random.randn(1)
c3 = np.random.randn(1)

# Hyperparameters
learningrate = 1e-9
numiter = 50000

for i in range(numiter):
    # forward pass
    y_hat=forward_pass(P_3,m3,c3)
    # Loss
   
    # Gradient
    dm = (1.0/N_3) *  -np.sum(np.multiply(P_3,(Vap_sat_3 - (m3*P_3 +c3))))
   
    dc = (1.0/N_3) * -np.sum(Vap_sat_3 - (m3*P_3 + c3))
    
    m3 = m3 -learningrate * dm
    c3 = c3 - learningrate * dc
    if i%5000==0:
        L_3.append(loss(y_hat,Vap_sat_3))
        print ("no of iterations:",i )
        plt.plot(P_3,Vap_sat_3,'go')
        plt.plot(P_3,y_hat,'r*')
        plt.ylabel('Data')
        plt.show()
        
R_3 = R2_score(P_3, y_hat)
```

```python
from sklearn.linear_model import LinearRegression
regressor_1 = LinearRegression()
regressor_1.fit(P_1, Vap_sat_1)
plt.scatter(P_1, Vap_sat_1, color = 'red')
plt.plot(P_1, regressor_1.predict(P_1), color = 'blue')
plt.ylabel('Pressure')
plt.show()
print(regressor_1.coef_[0],regressor_1.intercept_)
```
