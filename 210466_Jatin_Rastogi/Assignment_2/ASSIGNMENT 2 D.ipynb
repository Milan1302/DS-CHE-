```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
rng = np.random.RandomState(10)
x = 10 * rng.rand(50)
y = (10 * x**2 + 5 + 10*rng.randn(50)*x)/700
plt.scatter(x, y);
y
```

```python
x.shape
```

```python
N=x.shape[0]
N
L=[]
x=np.float128(x)
```

```python
def result(a,b,c,x):
    #print((a*(x**2)+b*x+c))
    return (a*(x**2)+b*x+c)
```

```python
def loss(y,y_hat):
    l=abs((y-y_hat)**7)
    
    return l.mean()
```

```python
a = np.random.randn(1)/300
b = np.random.randn(1)/300
c =np.random.randn(1)/300
a,b,c
```

```python
learning_rate = 0.000001
total_iter = 50001
```

```python
for i in range(total_iter):
    
    y_hat=result(a,b,c,x)
    da=(7/N)*np.sum(x**2*(a*x**2+b*x+c-y)*abs((a*(x**2)+b*x+c-y)**5))
    db=(7/N)*np.sum(x*(a*x**2+b*x+c-y)*abs((a*(x**2)+b*x+c-y)**5))
    dc=(7/N)*np.sum((a*x**2+b*x+c-y)*abs((a*(x**2)+b*x+c-y)**5))
    a=a-da*learning_rate
    b=b-db*learning_rate
    c=c-dc*learning_rate
    if i%10000==0:
        L.append(loss(y,y_hat))
        print("total iteration: ",i)
        plt.plot(x,y_hat,"b*")
        plt.plot(x,y,"go")
        
        plt.show()
```
