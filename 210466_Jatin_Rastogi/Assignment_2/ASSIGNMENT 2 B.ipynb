```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
rng = np.random.RandomState(10)
x = 10 * rng.rand(50)
y = (2 * x - 5 + rng.randn(50))
plt.scatter(x, y);
```

```python
x.shape
```

```python
N=x.shape[0]
N
L=[]
```

```python
def result(m1,x,c1):
    return m1*x+c1
```

```python
def loss(y,y_hat):
    l=abs((y-y_hat))
    return l.mean()
```

```python
m1 = np.random.randn(1)
c1 = np.random.randn(1)
m1,c1
```

```python
learning_rate = 0.001
total_iter = 50000
```

```python
for i in range(total_iter):
    
    y_hat=result(m1,x,c1)
    dm=1/N*np.sum(x*((m1*x+c1)-y))
    dc=1/N*np.sum(((m1*x+c1)-y))
    m1=m1-dm*learning_rate
    c1=c1-dc*learning_rate
    if i%10000==0:
        L.append(loss(y,y_hat))
        print("total iteration: ",i)
        plt.plot(x,y_hat,"r*")
        plt.plot(x,y,"go")
        
        plt.show()
```
