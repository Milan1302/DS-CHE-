# Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
rng=np.random.RandomState()
x=10*rng.rand(50)
y=4*x-5+rng.randn(50)
plt.scatter(x,y)
```

# Quadratic Function

```python
rng = np.random.RandomState()
x_q = 10 * rng.rand(50)
y_q = x_q**2 - 2*x_q + 3 + rng.randn(50)
plt.scatter(x_q, y_q);
```

```python
def loss_a(y_hat,y):
    l = np.abs((y_hat - y)**3/3) #cancels 3
    return l.mean()
```

```python
def loss_b(y_hat,y):
    l = np.abs((y_hat - y))
    return l.mean()
```

```python
def loss_c(y_hat,y):
    l = np.abs((y_hat - y)**4/4) #cancels 4
    return l.mean()
```

```python
def loss_d(y_hat,y):
    l = np.abs((y_hat - y)**7/7) #cancels 7
    return l.mean()
```

```python
l_a=[]
l_b=[]
l_c=[]
l_d=[]
N=x.shape[0]
```

# Forward Pass

```python
def forward_pass(x,m,c):
    return m*x+c

def forward_pass_q(x,a,b,d):
    return a*(x**2)+b*x+d
```

```python
m = np.random.randn(1)
a = np.random.randn(1)
b = np.random.randn(1)
m,a,b
```

```python
# Random Initialisation
m1 = np.random.randn(1)
c1 = np.random.randn(1)

# Hyperparameters
learningrate = 0.0000005
numiter = 50000

for i in range(numiter):
    

    # forward pass
    y_hat=forward_pass(x,m1,c1) 
    
    # gradient
    dm = (1.0/N) *  -np.sum(np.multiply(x,(y - (m1*x +c1))**2)) 
    dc = (1.0/N) * -np.sum((y - (m1*x +c1))**2)
    
    m1 = m1 -learningrate * dm
    c1 = c1 - learningrate * dc
                           
    if i%5000==0:
        l_a.append(loss_a(y_hat,y))
        print ("no of iterations:",i )
        plt.plot(x,y,'go')
        plt.plot(x,y_hat,'r*')
    #plt.plot(x,m1*x+c1,'r*')
        plt.ylabel('Data')
        plt.show()
```

```python
# Random Initialisation
m2 = np.random.randn(1)
c2 = np.random.randn(1)

# Hyperparameters
learningrate = 0.00002
numiter = 50000

for i in range(numiter):
    # forward pass
    y_hat_2=forward_pass(x,m2,c2) 
    
    # Gradient
    dm = (1.0/N) *  -np.sum(x)  
    dc = (1.0/N) * -np.sum(1)
    
    m2 = m2 -learningrate * dm
    c2 = c2 - learningrate * dc
                           
    if i%5000==0:
        l_b.append(loss_b(y_hat_2,y))
        print ("no of iterations:",i )
        plt.plot(x,y,'go')
        plt.plot(x,y_hat_2,'r*')
    #plt.plot(x,m1*x+c1,'r*')
        plt.ylabel('Data')
        plt.show()
```

```python
# Random Initialisation
a1 = np.random.randn(1)
b1 = np.random.randn(1)
d1 = np.random.randn(1)

# Hyperparameters
learningrate = 0.0000000007
numiter = 50000

for i in range(numiter):
    # forward pass
    y_hat_3=forward_pass_q(x_q,a1,b1,d1)
    
   
    # Gradient
    da = (1.0/N) *  -np.sum(np.multiply(x_q**2,(y_q - (a1*x_q**2 + b1*x_q +d1))**3))  
    db = (1.0/N) *  -np.sum(np.multiply(x_q,(y_q - (a1*x_q**2 + b1*x_q +d1))**3))
    dd = (1.0/N) *  -np.sum((y_q - (a1*x_q**2 + b1*x_q +d1))**3)
    
    a1 = a1 - learningrate * da
    b1 = b1 - learningrate * db
    d1 = d1 - learningrate * dd

    if i%5000==0:
        l_c.append(loss_c(y_hat_3,y_q))
        print ("no of iterations:",i )
        plt.plot(x_q,y_q,'go')
        plt.plot(x_q,y_hat_3,'r*')
    #plt.plot(x,m1*x+c1,'r*')
        plt.ylabel('Data')
        plt.show()
```
