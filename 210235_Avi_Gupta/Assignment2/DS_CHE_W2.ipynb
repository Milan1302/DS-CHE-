```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
r = np.random.RandomState(10)
x = 10 * r.rand(50)
yl = 2 * x - 5 + r.randn(50)
yp = (5*(x**2) + 2 * x - 5 + r.randn(50))
plt.plot(x, yp,'go');
plt.plot(x, yl,'ro');
```

```python
def loss1(yhat , y):
    l = ((np.abs(y-yhat))**3)/3
    return l.mean()
```

```python
def loss2(yhat , y):
    l = (np.abs(y-yhat))
    return l.mean()
```

```python
def loss3(yhat , y):
    l = ((np.abs(y-yhat))**4)/4
    return (l.mean())
```

```python
def loss4(yhat , y):
    l = ((np.abs(y-yhat))**7)/7
    return l.mean()
```

```python
N=x.shape[0]
```

```python
def linear(x,m,c):
    return m*x + c
```

```python
def poly(x,m1,m2,c):
    return (m1*(x**2) + m2*x + c)/4
```

```python
L1=[]
m1 = np.random.randn(1)
c1 = np.random.randn(1)

learningrate = 0.01
numiter = 50000

for i in range(numiter):
    
    y_hat=linear(x,m1,c1)
   
    dm = (2.0/N) *  -np.sum(np.multiply(x,(yl - (m1*x +c1))))
   
    dc = (2.0/N) * -np.sum(yl - (m1*x + c1))
    
    m1 = m1 -learningrate * dm
    c1 = c1 - learningrate * dc
    if i%5000==0:
        L1.append(loss1(y_hat,yl))
        print ("no of iterations:",i )
        plt.plot(x,yl,'go')
        plt.plot(x,y_hat,'r*')
        plt.ylabel('Data')
        plt.show()
```

```python
plt.plot(L1)
```
