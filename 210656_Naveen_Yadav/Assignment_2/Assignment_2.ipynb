```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Generating dataset for linear regression
rng=np.random.RandomState(seed=7)
x=rng.rand(50)
y=2*x-.9 +rng.randn(50)/10
plt.scatter(x,y);
```

```python
def cost_func(y_hat,y):
    cost=abs((y-y_hat)**3)/3
    return cost.mean()
```

```python
def for_pass(x,m,c):
    return x*m+c
```

```python
C=[]
N=x.shape[0]
```

```python
#For part (a)

#Initial Random initialization for m and c
m1=np.random.randn(1)
c1=np.random.randn(1)

#Hyperparameter Initialization
alpha=0.0025   #learning rate
epoch=50001    #number of iterations

for _ in range(epoch):
    y_hat=for_pass(x,m1,c1)
    C.append(cost_func(y_hat,y))
    
    #Gradient
    delta_m=(3.0/N)*np.sum(x*(y_hat-y)*abs(y_hat-y))
    delta_c=(3.0/N)*np.sum(np.multiply((y_hat-y),abs(y_hat-y)))
    
    #Updating values
    m1=m1-alpha*delta_m
    c1=c1-alpha*delta_c
    
    if _%5000==0:
        print ("No of iterations:",_ )
        plt.plot(x,y,'go')
        plt.plot(x,y_hat,'r*')
        plt.ylabel('Data')
        plt.show()
        
print(m1,c1)
```

```python
plt.plot(C)
```

```python
C_b=[]
N_b=x.shape[0]
```

```python
#For part (b)

#Initial Random initialization for m and c
m2=np.random.randn(1)
c2=np.random.randn(1)

#Hyperparameter Initialization
alpha=0.001   #learning rate
epoch=50001   #number of iterations

for _ in range(epoch):
    y_hat_b=for_pass_b(x,m2,c2)
    C_b.append(cost_func_b(y_hat_b,y))
    
    #Gradient
    delta_m_b=(1.0/N_b)*np.sum(x*(m2*x+c2-y)/abs(m2*x+c2-y))
    delta_c_b=(1.0/N_b)*np.sum((m2*x+c2-y)/abs(m2*x+c2-y))
    
    #Updating values
    m2=m2-alpha*delta_m_b
    c2=c2-alpha*delta_c_b
    
    if _%5000==0:
        print ("No of iterations:",_ )
        plt.plot(x,y,'go')
        plt.plot(x,y_hat_b,'r*')
        plt.ylabel('Data')
        plt.show()
print(m2,c2)
```

```python
#generating dataset for 2nd degree polynomial regression
rang=np.random.RandomState(seed=7)
x1=rang.rand(50)
y1= 3 * (x1 ** 2) - 2*x1 + rang.randn(50)/10 
plt.scatter(x1,y1);
```

```python
def loss_func_1(y_hat_c,y1):
    loss=abs((y_hat_c-y1)**4)/4
    return loss.mean()
```

```python
def forward_pass1(x1,a,b,c):
    return a*(x1**2)+b*x1+c
```

```python
L_c=[]
N_c=x1.shape[0]
```

```python
#For part (c)

#Initial Random initialization for m and c
a1=np.random.randn(1)/10
b1=np.random.randn(1)/10
c1=np.random.randn(1)/10
print(a1,b1,c1)
#Hyperparameter Initialization
alpha=0.5    #learning rate
epoch=50001  #number of iterations

for _ in range(epoch):
    y_hat_c=forward_pass1(x1,a1,b1,c1)
    L_c.append(loss_func_1(y_hat_c,y1))
    
    #Gradient
    delta_a_1=(4.0/N_c)*np.sum(np.multiply(x1**2,(a1*(x1**2)+b1*x1+c1-y1)**3))
    delta_b_1=(4.0/N_c)*np.sum(np.multiply(x1,(a1*(x1**2)+b1*x1+c1-y1)**3))
    delta_c_1=(4.0/N_c)*np.sum((a1*(x1**2)+b1*x1+c1-y1)**3)
    
    #Updating Values
    a1=a1-alpha*delta_a_1
    b1=b1-alpha*delta_b_1
    c1=c1-alpha*delta_c_1
    
    if _ % 5000==0:
        print ("No of iterations:",_ )
        plt.plot(x1,y1,'go')
        plt.plot(x1,y_hat_c,'r*')
        plt.ylabel('Data')
        plt.show()
print(a1,b1,c1)
```

```python
plt.plot(L_c)
```
