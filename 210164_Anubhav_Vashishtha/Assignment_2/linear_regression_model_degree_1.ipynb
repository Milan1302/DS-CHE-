```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
rng =np.random.RandomState(1)
x = rng.rand(50)
y =3*x+5+rng.rand(50)
plt.scatter(x,y)
x
```

```python
y
```

```python
def cost_func(y_hat,y):
    cost = (np.sum(np.absolute(y-y_hat)))/100
    return cost
```

```python
def gradient_descent_1(theta0, theta1,x,y):
    learning_rate =0.001;
    theta0 = theta0 -learning_rate/100*np.sum((theta0+theta1*x-y)/np.absolute(theta0+theta1*x-y))
    theta1 =theta1-learning_rate/100*np.sum((theta0+theta1*x-y)/np.absolute(theta0+theta1*x-y)*x);
    return theta0,theta1
    
```

```python
epochs =50000
L=[]
theta0=np.random.randn(1)
theta1=np.random.randn(1)
for i in range(epochs):
    y_h =  theta0+theta1*x;
    L.append(cost_func(y_h,y))
    theta0,theta1 =gradient_descent_1(theta0,theta1,x,y)
    if i%10000==0:
        plt.plot(x,y,'bo')
        plt.plot(x,y_h,'rD')
        plt.xlabel('epochs')
        plt.legend(['y','y_hat'])
        plt.ylabel('Data')
        plt.show()
     
```

```python
plt.plot(L,':')
```

```python
rng =np.random.RandomState(1)
x2 = rng.randn(50)

y2 =3*x2-5+rng.randn(50)
plt.scatter(x2,y2)
x2
```

```python
def cost_func2(y_hat,y):
    cost = (np.sum(np.power(np.absolute(y-y_hat),3)))/100
    return cost
```

```python
def gradient_descent_2(thta0, thta1,x2,y):
    learning_rate =0.0001;
    thta0 = thta0 -3*learning_rate/400*np.sum(np.square(thta0+thta1*x2-y))
    thta1 =thta1-3*learning_rate/400*np.sum(np.multiply(x2,np.square(thta0+thta1*x2-y)))
    return thta0,thta1
    
```

```python
epochs =50000
L1=[]
thta0=np.random.randn(1)/100
thta1=np.random.randn(1)/100
for i in range(epochs):
    y_h2 =  thta0+thta1*x2;
    L1.append(cost_func2(y_h2,y2))
    thta0,thta1 =gradient_descent_2(thta0,thta1,x2,y2)
   
    if i%10000==0:
        plt.plot(x2,y2,'bo')
        plt.plot(x2,y_h2,'rD')
        plt.xlabel('epochs')
        plt.legend(['y','y_hat'])
        plt.ylabel('Data')
        plt.show()
     
```

```python
rng =np.random.RandomState(1)
x = rng.rand(50)

y =3*(x**2)-2*x+rng.randn(50)/10
plt.scatter(x,y)
```

```python
def cost_func(y,y_h):
    cost =abs((y_h-y)**4)/4
    return cost.mean()
    
```

```python
def forward_pass(x1,a,b,c):
    return a*(x1**2)+b*x1+c
```

```python
L=[]
N=x.shape[0]
N
```

```python

epochs =50001
alpha =0.5
m0 = np.random.randn(1)/10
m1 = np.random.randn(1)/10
m2 = np.random.randn(1)/10
for i in range(epochs):
    y_h = m0+m1*x+m2*(x**2)
    L.append(cost_func(y,y_h))
    
    dm0 = (4/N)*np.sum((y_h-y)**3);
    dm1 = (4/N)*np.sum(np.multiply(x,(y_h-y)**3))
    dm2 = (4/N)*np.sum(np.multiply((x**2),(y_h-y)**3))
    m0 = m0-alpha*dm0
    m1 = m1-alpha*dm1
    m2 = m2-alpha*dm2
    if i%10000==0:
        plt.plot(x,y,'go')
        plt.plot(x,y_h,'r*')
        plt.show()
    
```
