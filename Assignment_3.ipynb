```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('superheated_vapor_properties.csv')
dataset
```

```python
V_data = dataset.loc[(dataset["Property"] == "V"),:]
V_data
```

```python
plt.plot(V_data['Pressure'],V_data['Vap_Sat'])
```

```python
def r2_score(y, y_hat):
    y_mean = np.mean(y)
    ss_tot = np.sum((y - y_mean)**2)
    ss_res = np.sum((y - y_hat)**2)
    return 1 - ss_res / ss_tot
```

```python
V_data_1 = V_data.loc[(V_data["Pressure"] < 300),:]
V_data_2 = V_data.loc[(V_data["Pressure"] >= 300) & (V_data["Pressure"] < 1500),:]
V_data_3 = V_data.loc[(V_data["Pressure"] >= 1500),:]
```

```python
plt.plot(V_data_1['Pressure'],V_data_1['Vap_Sat'])
```

```python
plt.plot(V_data_2['Pressure'],V_data_2['Vap_Sat'])
```

```python
plt.plot(V_data_3['Pressure'],V_data_3['Vap_Sat'])
```

```python
P_1 = V_data_1['Pressure'].values.reshape(-1,1)
Vap_sat_1 = V_data_1['Vap_Sat'].values.reshape(-1,1)
P_2 = V_data_2['Pressure'].values.reshape(-1,1)
Vap_sat_2 = V_data_2['Vap_Sat'].values.reshape(-1,1)
P_3 = V_data_3['Pressure'].values.reshape(-1,1)
Vap_sat_3 = V_data_3['Vap_Sat'].values.reshape(-1,1)
```

## Gradient descent approach

```python
def loss(y_hat, y):
    l = (y_hat - y) ** 2 / 2
    return l.mean()
```

```python
L_1=[]
L_2=[]
L_3=[]
N_1=P_1.shape[0]
N_2=P_2.shape[0]
N_3=P_3.shape[0]
```

```python
def forward_pass(x,m,c):
    return m*x+c
```

```python
# Random Initialisation
m1 = np.random.randn(1)
c1 = np.random.randn(1)

# Hyperparameters
learningrate = 0.25*1e-4
numiter = 50000

for i in range(numiter):
    # forward pass
    y_hat=forward_pass(P_1,m1,c1)
    # Loss
   
    # Gradient
    dm = (2.0/N_1) *  -np.sum(np.multiply(P_1,(Vap_sat_1 - (m1*P_1 +c1))))
   
    dc = (2.0/N_1) * -np.sum(Vap_sat_1 - (m1*P_1 + c1))
    
    m1 = m1 -learningrate * dm
    c1 = c1 - learningrate * dc
    if i%5000==0:
        L_1.append(loss(y_hat,Vap_sat_1))
        print ("no of iterations:",i )
        plt.plot(P_1,Vap_sat_1,'go')
        plt.plot(P_1,y_hat,'r*')
        plt.ylabel('Data')
        plt.show()

R_1 = r2_score(P_1, y_hat)
```

```python
# Random Initialisation
m2 = np.random.randn(1)
c2 = np.random.randn(1)

# Hyperparameters
learningrate = 0.1*1e-6
numiter = 50000

for i in range(numiter):
    # forward pass
    y_hat=forward_pass(P_2,m2,c2)
    # Loss
   
    # Gradient
    dm = (1.0/N_2) *  -np.sum(np.multiply(P_2,(Vap_sat_2 - (m2*P_2 +c2))))
   
    dc = (1.0/N_2) * -np.sum(Vap_sat_2 - (m2*P_2 + c2))
    
    m2 = m2 -learningrate * dm
    c2 = c2 - learningrate * dc
    if i%5000==0:
        L_2.append(loss(y_hat,Vap_sat_2))
        print ("no of iterations:",i )
        plt.plot(P_2,Vap_sat_2,'go')
        plt.plot(P_2,y_hat,'r*')
        plt.ylabel('Data')
        plt.show()
        
R_2 = r2_score(P_2, y_hat)
```

```python
# Random Initialisation
m3 = np.random.randn(1)
c3 = np.random.randn(1)

# Hyperparameters
learningrate = 1e-8
numiter = 50000

for i in range(numiter):
    # forward pass
    y_hat=forward_pass(P_3,m3,c3)
    # Loss
   
    # Gradient
    dm = (1.0/N_3) *  -np.sum(np.multiply(P_3,(Vap_sat_3 - (m3*P_3 +c3))))
   
    dc = (1.0/N_3) * -np.sum(Vap_sat_3 - (m3*P_3 + c3))
    
    m3 = m3 -learningrate * dm
    c3 = c3 - learningrate * dc
    if i%5000==0:
        L_3.append(loss(y_hat,Vap_sat_3))
        print ("no of iterations:",i )
        plt.plot(P_3,Vap_sat_3,'go')
        plt.plot(P_3,y_hat,'r*')
        plt.ylabel('Data')
        plt.show()

R_3 = r2_score(P_3, y_hat)
```
