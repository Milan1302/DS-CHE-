```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
rng = np.random.RandomState()
x = 10 * rng.rand(50)
```

```python
rng = np.random.RandomState()
x_q = 10 * rng.rand(50)
y_q = x_q**2 - 5*x_q - 6 + rng.randn(50)
plt.scatter(x_q, y_q);
```

```python
def loss_3(y_hat, y):
    n = (y_hat - y)**4/4
    return n.mean()
```

```python
def loss_4(y_hat, y):
    o = np.abs((y_hat - y)**7)/7 
    return o.mean()
```

```python
L_3=[]
L_4=[]
N=x.shape[0]
```

```python
def forward_pass_q(x,a,b,e):
    return a*x**2 + b*x + e
```

```python
m = np.random.randn(1)
a = np.random.randn(1)
b = np.random.randn(1)
m,a,b
```

```python
a1 = np.random.randn(1)
b1 = np.random.randn(1)
c1 = np.random.randn(1)

learningrate = 0.0000000007
numiter = 50000

for i in range(numiter):
    y_hat_3=forward_pass_q(x_q,a1,b1,c1)
    da = (1.0/N) *  -np.sum(np.multiply(x_q**2,(y_q - (a1*x_q**2 + b1*x_q +c1))**3))
    db = (1.0/N) *  -np.sum(np.multiply(x_q,(y_q - (a1*x_q**2 + b1*x_q +c1))**3))
    dc = (1.0/N) *  -np.sum((y_q - (a1*x_q**2 + b1*x_q +c1))**3)
    
    a1 = a1 - learningrate * da
    b1 = b1 - learningrate * db
    c1 = c1 - learningrate * dc

    if i%5000==0:
        L_3.append(loss_3(y_hat_3,y_q))
        print ("no of iterations:",i )
        plt.plot(x_q,y_q,'go')
        plt.plot(x_q,y_hat_3,'r*')
        plt.ylabel('Data')
        plt.show()
```

```python
a2 = np.random.randn(1)
b2 = np.random.randn(1)
c2 = np.random.randn(1)

learningrate = 0.0000000000001
numiter = 50000

for i in range(numiter):
    y_hat_4=forward_pass_q(x_q,a2,b2,c2)
    
    da = (1.0/N) *  -np.sum(np.multiply(x_q**2,(y_q - (a2*x_q**2 + b2*x_q +c2))**6))
    db = (1.0/N) *  -np.sum(np.multiply(x_q,(y_q - (a2*x_q**2 + b2*x_q +c2))**6))
    dc = (1.0/N) *  -np.sum((y_q - (a2*x_q**2 + b2*x_q +c2))**6)
    
    a2 = a2 - learningrate * da
    b2 = b2 - learningrate * db
    c2 = c2 - learningrate * dc

    if i%5000==0:
        L_4.append(loss_4(y_hat_4,y_q))
        print ("no of iterations:",i )
        plt.plot(x_q,y_q,'go')
        plt.plot(x_q,y_hat_4,'r*')
        plt.ylabel('Data')
        plt.show()
```
