## Using sklearn 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
df= pd.read_csv('superheated_vapor_properties.csv')
df
```

```python
v_data=df[df['Property']=='V']
v_data
```

```python
df_low = v_data[v_data["Pressure"] < 300][['Pressure','Property','Liq_Sat']]
df_low
```

```python
df_medium = v_data[(v_data["Pressure"] >= 300) & (v_data["Pressure"] < 1500)][['Pressure','Property','Liq_Sat']]
df_medium
```

```python
df_high = v_data[v_data["Pressure"] >= 1500][['Pressure','Property','Liq_Sat']]
df_high
```

```python
# linear regression model for the pressure range <300 kPa
from sklearn.linear_model import LinearRegression
X_low = df_low["Pressure"].values.reshape(-1, 1)  # use pressure as the feature
y_low = df_low["Liq_Sat"].values
model_low = LinearRegression()
model_low.fit(X_low, y_low)
predictions_low = model_low.predict(X_low)
# Calculate the R2 score, slope and intercept for the low pressure range model
r2_low = model_low.score(X_low, y_low)
slope_low = model_low.coef_[0]
intercept_low = model_low.intercept_
# Plot 
plt.figure(figsize=(10,10))
plt.scatter(X_low, y_low,color='r',s=15)
plt.plot(X_low, predictions_low, color='c')
plt.title("Low Pressure Range (R2 = {:.2f})".format(r2_low))
plt.xlabel("Pressure (kPa)")
plt.ylabel("Specific Volume (m^3/kg)")
plt.show()
print("Slope =",slope_low)
print("Intercept =",intercept_low)
print("R2 =",r2_low)
print(" ")
```

```python
# linear regression model for the pressure range [300,1500]
X_medium = df_medium["Pressure"].values.reshape(-1, 1)  # use pressure as the feature
y_medium = df_medium["Liq_Sat"].values
model_medium = LinearRegression()
model_medium.fit(X_medium, y_medium)
predictions_medium = model_medium.predict(X_medium)

# Calculate the R2 score, slope and intercept for the medium pressure range model
r2_medium = model_medium.score(X_medium, y_medium)
slope_medium = model_medium.coef_[0]
intercept_medium = model_medium.intercept_

# Plot 
plt.figure(figsize=(10,10))
plt.scatter(X_medium, y_medium,s=15)
plt.plot(X_medium, predictions_medium, color='red')
plt.title("Medium Pressure Range (R2 = {:.2f})".format(r2_medium))
plt.xlabel("Pressure (kPa)")
plt.ylabel("Specific Volume (m^3/kg)")
plt.show()
print("Slope =",slope_medium)
print("Intercept =",intercept_medium)
print("R2 =",r2_medium)
print(" ")
```

```python
# linear regression model for the medium >1500kPa
X_high = df_high["Pressure"].values.reshape(-1, 1)  # use pressure as the feature
y_high = df_high["Liq_Sat"].values
model_high = LinearRegression()
model_high.fit(X_high, y_high)
predictions_high = model_high.predict(X_high)

# Calculate the R2 score, slope and intercept for the medium pressure range model
r2_high = model_high.score(X_high, y_high)
slope_high = model_high.coef_[0]
intercept_high = model_high.intercept_

# Plot
plt.figure(figsize=(10,10))
plt.scatter(X_high, y_high,s=15)
plt.plot(X_high, predictions_high, color='red')
plt.title("Medium Pressure Range (R2 = {:.2f})".format(r2_high))
plt.xlabel("Pressure (kPa)")
plt.ylabel("Specific Volume (m^3/kg)")
plt.show()
print("Slope =",slope_high)
print("Intercept =",intercept_high)
print("R2 =",r2_high)
print(" ")
```

## Using Scratch Implementation


```python
import matplotlib.pyplot as plt
sv_low = df_low['Liq_Sat']
p_low = df_low['Pressure']
sv_medium = df_medium['Liq_Sat']
p_medium = df_medium['Pressure']
sv_high = df_high['Liq_Sat']
p_high = df_high['Pressure']
for sv, p in [(sv_low, p_low), (sv_medium, p_medium), (sv_high, p_high)]:
    # Calculate the mean and standard deviation of the specific volume and pressure
    mean_sv = np.mean(sv)
    std_sv = np.std(sv)
    mean_p = np.mean(p)
    std_p = np.std(p)

    # Calculate the slope and intercept of the linear regression
    slope = np.sum((p - mean_p) * (sv - mean_sv)) / np.sum((p - mean_p)**2)
    intercept = mean_sv - slope * mean_p

    # Calculate the R2 score
    r2 = 1 - np.sum((sv - (slope * p + intercept))**2) / np.sum((sv - mean_sv)**2)
    print("Slope=",slope)
    print("Intercepr=",intercept)
    print("R2=",r2)
    plt.figure(figsize=(10,10))
    # Plot the data and the linear regression line
    plt.scatter(p, sv, label='Data',c='c',s=15)
    plt.plot(p, slope * p + intercept, color='red', label='Linear Regression')
    plt.title(f'R2 Score: {r2:.2f}')
    plt.xlabel('Pressure')
    plt.ylabel('Specific Volume')
    plt.show()
```
