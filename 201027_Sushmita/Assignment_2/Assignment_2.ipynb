# Linear Regression 


```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

### Linear Function

```python
import numpy as np

# Set the range of x values
x_min = 0
x_max = 10

# Generate a set of random x values
x = np.random.uniform(x_min, x_max, size=100)

# Choose the coefficients for the linear equation
a = 2
b = 1

# Calculate the corresponding y values
y = a * x + b

# Add random noise to the y values
y += np.random.normal(scale=0.5, size=100)
plt.figure(figsize=(15,10))
plt.scatter(x,y,s=20,color='k')
plt.xlabel("X")
plt.ylabel("Y")
```

### Quadratic Function

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the range of x values
x_min = 0
x_max = 10

# Generate a set of random x values
x_q = np.random.uniform(x_min, x_max, size=100)

# Choose the coefficients for the quadratic equation
a = 1
b = 2
c = 3

# Calculate the corresponding y values
y_q = a * x_q**2 + b * x_q + c

# Add random noise to the y values
y_q += np.random.normal(scale=0.5, size=100)

# Create a figure and plot the scatter plot
plt.figure(figsize=(15,10))
plt.scatter(x_q, y_q, s=20, color='c')

# Add labels to the x and y axes
plt.xlabel("X")
plt.ylabel("Y")

# Display the plot
plt.show()
```

### Loss Functions For Linear Regression Model 

```python
def loss_a(y_hat, y):
    a= np.abs((y_hat - y)**3/3) #division by 3 cancels out the power of 3 in the loss function.
    return a.mean()
```

```python
def loss_b(y_hat, y):
    b= np.abs((y_hat - y)) 
    return b.mean()
```

### Forward pass function


```python
def forward_pass_Linear(x, w, b):
  y = w * x + b
  return y
```

```python
N=x.shape[0]
N
```

### Gradient Descent Algorithm

```python
import matplotlib.pyplot as plt

# Set the initial values for the model parameters
w = np.random.randn(1)
b = np.random.randn(1)

# Set the learning rate
learning_rate = 0.00005

# Set the number of iterations
num_iterations = 60000

# Initialize lists to store the loss and parameter values during the iterations
losses = []
ws = []
bs = []

# Loop over the number of iterations
for i in range(num_iterations):
  # Calculate the forward pass
  y_hat = forward_pass_Linear(x, w, b)

  # Calculate the loss
  l = loss_a(y_hat, y)

  # Calculate the gradients of the loss with respect to the parameters
  dw = 3 * np.sum(np.abs(y_hat - y) ** 2 * np.sign(y_hat - y) * x) / y.shape[0]
  db = 3 * np.sum(np.abs(y_hat - y) ** 2 * np.sign(y_hat - y)) / y.shape[0]
#   # Update the parameters using gradient descent
  w = w -learning_rate * dw
  b = b - learning_rate * db
  if i%5000==0:
      print ("No of iterations",i)
      plt.figure(figsize=(5,5))
      plt.plot(x,y,'m.',markersize=7)
      plt.plot(x,y_hat,'b',markersize=7)
      plt.ylabel('Data')
      plt.show()

  # Store the loss and parameter values during the iterations
  losses.append(l)
  ws.append(w)
  bs.append(b)

# Plot the loss over the iterations
plt.plot(losses)
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.show()

# Print the final values of the parameters
print("w=",w,"b=",b)
```

```python
def loss_c(y_hat, y):
    c= (y_hat - y)**4/4
    return c.mean()
```

```python
def loss_d(y_hat, y):
    d = np.abs((y_hat - y)**7)/7 
    return d.mean()
```

```python
def forward_pass_poly(x, w1, w2, b):
  y_hat = w1 * x**2 + w2 * x + b
  return y_hat
```
