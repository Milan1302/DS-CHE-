# 3)Quadratic regression model using (c.) as loss function

```python
def loss(y_hat, y):
    diff=( y_hat-y)**4
    abs_diff= np.absolute(diff)
    mean_diff= abs_diff.mean()
    return mean_diff
```

```python
def forward_pass(x,m,m2,c):
    return m*(x**2)+m2*x+c
```

```python
m1 = np.random.randn(1)
c1 = np.random.randn(1)
m2 = np.random.randn(1)

# Hyperparameters
learningrate = 0.001
numiter = 50000

for i in range(numiter):
    # forward pass
    y_hat=forward_pass(x,m1,m2,c1)
    
    # Gradient
    dm = (4.0/N) *  -np.sum(np.multiply((x**2),((y - (m*(x**2)+m2*x+c1))**3)))
    
    dm2 = (4.0/N) *  -np.sum(np.multiply(x,(y - (m*(x**2)+m2*x+c1))**3))
   
    dc = (4.0/N) * -np.sum(y - (m*(x**2)+m2*x+c1))
    
    m1 = m1 -learningrate * dm
    c1 = c1 - learningrate * dc
    m2 = m2 -learningrate * dm2
    if i%5000==0:
        L.append(loss(y_hat,y))
        print ("no of iterations:",i )
        plt.plot(x,y,'go')
        plt.plot(x,y_hat,'r*')
    #plt.plot(x,m1*x+c1,'r*')
        plt.ylabel('Data')
        plt.show()
```
