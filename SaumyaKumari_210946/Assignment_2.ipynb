```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline

rng = np.random.RandomState(10)
x = 10 * rng.rand(50)
y = ((x - np.mean(x))**2)**(4/2)
plt.scatter(x, y);
```

```python
def loss(y_hat, y):
    l = (y_hat - y) ** 2 / 2
    return l.mean()
```

```python
x.shape
```

```python
def forward_pass(x,a,b,c):
    return a*x*x+b*x+c
```

```python
# Random Initialisation
a1 = np.random.randn(1)
b1 = np.random.randn(1)
c1 = np.random.randn(1)

# Hyperparameters
learningrate = 0.01
N = 50000   #numiter

for i in range(N):
    # forward pass
    y_hat=forward_pass(x,a1,b1,c1)
    # Loss
   
    # Gradient
    da1 = (2.0/N) *  -np.sum(np.multiply(x**2,(y - (a1*x*x +b1*x +c1))))
    db1 = (2.0/N) *  -np.sum(np.multiply(x,(y - (a1*x*x +b1*x +c1))))
   
    dc1 = (2.0/N) * -np.sum(y - (a1*x*x +b1*x +c1))
    
    a1 = a1 -learningrate * da1
    b1 = b1 - learningrate * db1
    c1 = c1 - learningrate * dc1
    if i%5000==0:
        
        print ("no of iterations:",i )
        plt.plot(x,y,'go')
        plt.plot(x,y_hat,'r*')
    #plt.plot(x,a1*x**2+b1*x+c1,'r*')
        plt.ylabel('Data')
        plt.show()
```
